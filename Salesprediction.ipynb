# Assinging dependent and independent variables

```python
Data.corr()
```

```python
Data = Data.drop(labels=["Unnamed: 0"],axis=1)
```

```python
Data.head()
```

```python
Y = Data[["Sales"]]
X = Data[["TV","Radio","Newspaper"]]
```

# Split the data into training and testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,Y,random_state=20, test_size=0.25)
```

# Making Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(X_train,y_train)
```

# Predicting the output

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,pred)
```

# Making model with other random state

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,Y,random_state=20, test_size=0.20)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,pred)
```

# Checking model result for different different randomstate

```python
for i in range(10,100,5):
    from sklearn.model_selection import train_test_split
    X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=i)
    from sklearn.linear_model import LinearRegression
    lm = LinearRegression()
    model = lm.fit(X_train,y_train)
    pred = model.predict(X_test)
    from sklearn.metrics import mean_absolute_error
    print(i,mean_absolute_error(y_test,pred))
```
